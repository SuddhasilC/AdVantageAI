```python
customer_spend_df = pd.DataFrame(transactions, columns=['Credit_card', 'Selling_price'])
# Group by 'Credit_card' and calculate the sum of 'Selling_price'
customer_spend_df = customer_spend_df.groupby('Credit_card')['Selling_price'].sum().reset_index()
customer_spend_df = customer_spend_df.rename(columns={'Credit_card': 'C_ID'})
```

```python
min_price = customer_spend_df["Selling_price"].min()
max_price = customer_spend_df["Selling_price"].max()
total_range = max_price - min_price
range_width = total_range / 3
ranges = [min_price, min_price + range_width, min_price + 2 * range_width, max_price]
```

```python
customer_spend_df["Price_Range"] = pd.cut(customer_spend_df["Selling_price"], bins=ranges, labels=["low spend", "medium spend", "high spend"])
```

```python
low_spend_C_IDs = customer_spend_df[customer_spend_df["Price_Range"] == "low spend"]["C_ID"].tolist()
medium_spend_C_IDs = customer_spend_df[customer_spend_df["Price_Range"] == "medium spend"]["C_ID"].tolist()
high_spend_C_IDs = customer_spend_df[customer_spend_df["Price_Range"] == "high spend"]["C_ID"].tolist()
```

```python
max_length = max(len(low_spend_C_IDs), len(medium_spend_C_IDs), len(high_spend_C_IDs))
```

```python
low_spend_C_IDs += [None] * (max_length - len(low_spend_C_IDs))
medium_spend_C_IDs += [None] * (max_length - len(medium_spend_C_IDs))
high_spend_C_IDs += [None] * (max_length - len(high_spend_C_IDs))
```

```python
data_new = {
    "low spending customer (C_ID)": low_spend_C_IDs,
    "medium spending customer (C_ID)": medium_spend_C_IDs,
    "high spending customer (C_ID)": high_spend_C_IDs,
}
```

```python
def find_highest_5_spending_product(df):
    # Get the top 5 rows based on 'spend' column
    top_5_rows = df.nlargest(5, 'spend')
    # Reset the index to get a new default integer index
    top_5_rows = top_5_rows.reset_index(drop=True)
    return top_5_rows
highest_5_spending_product_df = find_highest_5_spending_product(grouped_df_by_product)
print(highest_5_spending_product_df.to_string())
```

```python
def find_highest_5_spending_state(df):
    # Get the top 5 rows based on 'spend' column
    top_5_rows = df.nlargest(5, 'spend')
    # Reset the index to get a new default integer index
    top_5_rows = top_5_rows.reset_index(drop=True)
    return top_5_rows
highest_5_spending_state_df = find_highest_5_spending_state(grouped_df_by_state)
print(highest_5_spending_state_df.to_string())
```

```python
def find_highest_5_spending_payment_method(df):
    # Get the top 5 rows based on 'spend' column
    top_5_rows = df.nlargest(5, 'spend')
    # Reset the index to get a new default integer index
    top_5_rows = top_5_rows.reset_index(drop=True)
    return top_5_rows
highest_5_spending_payment_method_df = find_highest_5_spending_payment_method(grouped_df_by_payment_method)
print(highest_5_spending_payment_method_df.to_string())
```

```python
df_encoded = pd.get_dummies(X, columns=['Gender'], prefix='Category')
```

```python
X['Age']=scaler.fit_transform(X[['Age']])
```

```python
y=tt[['Price ','P_CATEGORY','CONDTION']]
```

```python
df_encoded = pd.get_dummies(y, columns=['P_CATEGORY'], prefix='P')
```

```python
df_encoded = pd.get_dummies(df_encoded, columns=['CONDTION'], prefix='Category')
```

```python
y
```

```python
y=df_encoded
```

```python
outCol=y.columns
```

```python
y
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
# Define the model
model = Sequential()

# Input layer
model.add(Dense(128, input_dim=28, activation='tanh'))

# Hidden layers
model.add(Dense(256, activation='tanh'))
model.add(Dropout(0.3))  # Reduced dropout rate

model.add(Dense(128, activation='tanh'))
model.add(Dropout(0.6))  # Reduced dropout rate

model.add(Dense(64, activation='tanh'))

# Output layer for classification with softmax activation
model.add(Dense(17, activation='softmax'))  # 17 classes

# Compile the model with categorical cross-entropy loss and accuracy metric
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Print model summary
model.summary()

# Train the model with more epochs
history = model.fit(X, y, epochs=50, batch_size=32, validation_split=0.2)

# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(X, y)
print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)
```
